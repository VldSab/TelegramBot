```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import chardet
from sklearn.cluster import KMeans
import numpy as np
```

```python
name = 'clean_zaprosy.csv'

with open(name, 'rb') as f:
    result = chardet.detect(f.read())  # or readline if the file is large

df = pd.read_csv(name, encoding = result['encoding'], delimiter = ',')

phrases = df['Запрос']
phrases = phrases.to_list()
```

```python
class LemmTFIDFvector(TfidfVectorizer):
    def analyzer(self):
        analyzer = super(TfidfVectorizer, self).build_analyzer()
        return lambda x: (m.parse(word)[0].normal_form for word in analizer(x))
    
def KM_fit(phrases, n_clusters = 100, init = 'k-means++', n_init = 10, max_iter = 300, tol = 0.0001, n_jobs = 4, min_df = 2, stop_words = None):
    tfidf_v = LemmTFIDFvector(min_df = min_df, stop_words = stop_words) # есть параметр stop_words = [*список строк, которые выкидываем*]
    X = tfidf_v.fit_transform(phrases)
    k_means = KMeans(n_clusters = n_clusters, init = init, n_init = n_init, max_iter = max_iter, tol = tol, n_jobs = n_jobs)
    k_means.fit(X)
    return k_means.labels_

def TFID_clustering(n_clusters, phrases, filename, min_df = 2, stop_words = None):
    labels = KM_fit(phrases, n_clusters = n_clusters, min_df = min_df, stop_words = stop_words)
    dct = {}
    for k, label in zip(phrases, k_means.labels_):
        dct[label] = dct.get(label, []) + [k]
    with open(filename, 'w') as f:
        for c in dct:
            for k in dct[c]:
                f.write('{};{}\n'.format(c, k))    
```
