```python
import pandas as pd
import numpy as np
```

```python
def script(df, x_word):
    return df[df.Запрос.str.find(x_word) != -1]

def script_minus(df, xm_word):
    df1 = df[df.Запрос.str.find(xm_word) != -1]
    return pd.concat([df, df1]).drop_duplicates(keep=False)

def script_cpc(df, x_value):
    df['СPC Yandex'] = [x.replace(',', '.') if type(x) is not float else x for x in df['СPC Yandex'] ]
    df['СPC Yandex'] = df['СPC Yandex'].astype(float)
    return df[df['СPC Yandex'] <= x_value]

def main(path, x_word = '', xm_word = '', x_value = np.nan):
    df = pd.read_csv(path, header = 0, delimiter= ';')
    if x_word != '':
        df = script(df, x_word)
    if xm_word != '':
        df = script_minus(df, xm_word)
    if x_value is not np.nan:
        df = script_cpc(df, x_value)
        
    df.to_csv("output.csv")
    return df
```
